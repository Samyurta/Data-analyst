```python
# gender and product_category wise maximum amount
data1=data.groupby(['Gender','Product_Category'])['Amount'].sum().sort_values(ascending=False).reset_index()
data1
```

```python
sns.barplot(x='Gender', y='Amount', data=data1, hue='Product_Category')
```

```python
data.groupby(['Gender', 'State', 'Product_Category'])['Amount'].sum().reset_index()
```

```python
data.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).reset_index()
```

```python
data1=data.groupby(['State', 'Gender'])['Amount'].sum().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize=(15, 5))
sns.barplot(x='State', y='Amount', data=data1, hue='Gender')
plt.xticks(rotation=75)
plt.show()
```
